```python
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题


def dx(r,x,a,y,t):
    return r*x - a*x*y

def dy(d,x,b,y,t):
    return -d*y + b*x*y

def forward(r,x_0,a,y_0,b,d,t):
    X = []
    Y = []
    X.append(x_0)
    Y.append(y_0)
    for i in range(0,len(t)-1):
        X.append( X[i-1] + dx(r,X[i-1],a,Y[i-1],t[i]) * (t[i+1] - t[i]) )
        Y.append( Y[i-1] + dy(d,X[i-1],b,Y[i-1],t[i]) * (t[i+1] - t[i]) )

    return X,Y,t

r = 1
a = 0.1
b = 0.02
d = 0.5
x_0 = 25
y_0 = 2
t = np.linspace(0, 30, 20000)


X1,Y1,t = forward(r,x_0,a,y_0,b,d,t)
plt.plot(t,X1,label = 'X1')
plt.plot(t,Y1,label = 'Y1')
title = f'X1 和 Y1 随时间变化曲线 (x_0 = {x_0}, y_0 = {y_0})'
plt.title(title)
plt.legend()
plt.show()

X2,Y2,t = forward(r,x_0*5,a,y_0*5,b,d,t)
plt.plot(t,X2,label = 'X2')
plt.plot(t,Y2,label = 'Y2')
title = f'X2 和 Y2 随时间变化曲线 (x_0 = {x_0*5}, y_0 = {y_0*5})'
plt.title(title)
plt.legend()
plt.show()

X3,Y3,t = forward(r,x_0*10,a,y_0*10,b,d,t)
plt.plot(t,X3,label = 'X3')
plt.plot(t,Y3,label = 'Y3')
title = f'X3 和 Y3 随时间变化曲线 (x_0 = {x_0*10}, y_0 = {y_0*10})'
plt.title(title)
plt.legend()
plt.show()


# 绘制散点图
cmap = plt.get_cmap('YlOrBr')
colors = [cmap(0.25), cmap(0.40), cmap(0.55)]

plt.scatter(X1, Y1, s=0.05, color=colors[0], label='(X1, Y1) 点')
plt.scatter(X2, Y2, s=0.05, color=colors[1], label='(X2, Y2) 点')
plt.scatter(X3, Y3, s=0.05, color=colors[2], label='(X3, Y3) 点')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('(X, Y) 点的分布')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义x的取值范围
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = np.zeros_like(x)
N = 10  # 取级数的前N项进行近似

for n in range(1, N + 1):
    term = (-1) ** (n - 1) * np.sin(n * x) / n
    y += term

# 绘制图像
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Plot of the series $\sum_{n=1}^{\infty} (-1)^{n-1} \frac{\sin nx}{n}$')
plt.grid(True)
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题


def dx(r,k,x,a,y,t):
    return r*x - a*x*y - r/k * x**2

def dy(d,m,x,b,y,t):
    return -d*y + b*x*y - m*y**2 

def forward(r,x_0,a,y_0,b,d,k,m,t):
    X = []
    Y = []
    X.append(x_0)
    Y.append(y_0)
    for i in range(0,len(t)-1):
        X.append( X[i-1] + dx(r,k,X[i-1],a,Y[i-1],t[i]) * (t[i+1] - t[i]) )
        Y.append( Y[i-1] + dy(d,m,X[i-1],b,Y[i-1],t[i]) * (t[i+1] - t[i]) )

    return X,Y,t

r = 1
a = 0.1
b = 0.02
d = 0.5
k = 1000
m = 0.1
x_0 = 25
y_0 = 2
t = np.linspace(0, 30, 20000)


X1,Y1,t = forward(r,x_0,a,y_0,b,d,k,m,t)
plt.plot(t,X1,label = 'X1')
plt.plot(t,Y1,label = 'Y1')
title = f'X1 和 Y1 随时间变化曲线 (x_0 = {x_0}, y_0 = {y_0})'
plt.title(title)
plt.legend()
plt.show()

X2,Y2,t = forward(r,x_0*5,a,y_0*5,b,d,k,m,t)
plt.plot(t,X2,label = 'X2')
plt.plot(t,Y2,label = 'Y2')
title = f'X2 和 Y2 随时间变化曲线 (x_0 = {x_0*5}, y_0 = {y_0*5})'
plt.title(title)
plt.legend()
plt.show()

X3,Y3,t = forward(r,x_0*10,a,y_0*10,b,d,k,m,t)
plt.plot(t,X3,label = 'X3')
plt.plot(t,Y3,label = 'Y3')
title = f'X3 和 Y3 随时间变化曲线 (x_0 = {x_0*10}, y_0 = {y_0*10})'
plt.title(title)
plt.legend()
plt.show()


# 绘制散点图
cmap = plt.get_cmap('YlOrBr')
colors = [cmap(0.25), cmap(0.40), cmap(0.55)]

plt.scatter(X1, Y1, s=0.05, color=colors[0], label='(X1, Y1) 点')
plt.scatter(X2, Y2, s=0.05, color=colors[1], label='(X2, Y2) 点')
plt.scatter(X3, Y3, s=0.05, color=colors[2], label='(X3, Y3) 点')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('(X, Y) 点的分布')
plt.grid(True)
plt.legend()
plt.show()
```
